```python
import requests
import json
import pandas as pd

final_news_table = pd.DataFrame()
import datetime
now = datetime.datetime.now()  # Get the current datetime object

months = []
years = []

for i in range(240):  # Loop through 240 months (20 years)
    months.append(now.month)  # Add the month to the months list
    years.append(now.year)  # Add the year to the years list
    now -= datetime.timedelta(days=30)

months_and_years = pd.DataFrame(list(zip(months, years)), columns=['month','year'])

for i in months_and_years:
    month = months_and_years['month'][i]
    year = months_and_years['year'][i]
    
    requestUrl = f"https://api.nytimes.com/svc/archive/v1/{year}/{month}.json"
    parameters = {'q':"stock market",
                  'fq':'document_type: "article"',
                  'api-key': 'GjOWqEAGGtwRJyPAgaDDC90Zrlf8iYcA'
                 }
    response = requests.get(requestUrl, params=parameters).json()
    headlines = pd.DataFrame(response["response"]["docs"])[['headline', 'abstract', 'snippet', 'pub_date']]
    head = []
    for i in headlines['headline']:
        head.append(i['main'])

    headlines['headline'] = head
    final_news_table.append(headlines)

final_news_table.head(10)
```
